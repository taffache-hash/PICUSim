# ARDS PEEP titration

Educational/research alpha notebook. Not for clinical use. Not a medical device. Not a validated patient-specific digital twin.


```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if not (ROOT / 'run_simulation.py').exists():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

from core import ScenarioLoader
from run_simulation import build_twin
```

```python
scenario_path = ROOT / 'scenarios' / 'ards_mild.yaml'
loader = ScenarioLoader.from_yaml(scenario_path)
bus = loader.build_bus()
engine = build_twin(bus, loader.config, dt=1.0)
engine.add_perturbations(loader.build_perturbations())
df = engine.run(T=loader.simulation_time)
df.tail()
```

```python
cols = [c for c in ['SaO2','PaO2','PaCO2','MAP','HR','Vt','EELV','recruited_frac','MP'] if c in df.columns]
df[cols].plot(figsize=(12, 6), title=loader.scenario_name)
```

```python
final = df.iloc[-1]
for key in ['SaO2','PaO2','PaCO2','pH_a','MAP','HR','CO','Vt','EELV','recruited_frac','MP']:
    if key in final.index:
        print(f'{key:20s} {final[key]}')
```
